```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
mouse_metadata = "data\Mouse_metadata.csv"
study_results = "data\Study_results.csv"
```

```python
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)
mouse_metadata
```

```python
study_results
```

```python
# Combine the data into a single dataset
merged_data = pd.merge(mouse_metadata, study_results, how='left', on="Mouse ID")
merged_data.head()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean = merged_data.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
median = merged_data.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
variance = merged_data.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
std = merged_data.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
SEM = merged_data.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

                                                   
summary_table = pd.DataFrame({"Mean Tumor Volume": mean,
                             "Median Tumor Volume": median,
                             "Tumor Volume Variance ": variance,
                             "Tumor Volume Std": std,
                             "Tumor Volume SEM": SEM})      
summary_table
```

```python
# Generate a bar plot showing number of data points for each treatment regimen using pandas
point_counts = merged_data['Drug Regimen'].value_counts()
count_chart = point_counts.plot(kind='bar', color='c')
count_chart.set_xlabel("Drug Regimen")
count_chart.set_ylabel("Number of Points")
```

```python
# Generate a bar plot showing number of data points for each treatment regimen using pyplot
counts = merged_data['Drug Regimen'].value_counts()
x_axis = np.arange(len(counts))
plt.bar(x_axis, counts, color='b', alpha=0.5, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, merged_data["Drug Regimen"], rotation="vertical")
plt.xlabel('Drug Regimen')
plt.ylabel('Number of Points')
```
