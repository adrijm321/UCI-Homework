```python
import pandas_datareader.data as web
import pandas as pd
import numpy as np
import datetime as dt
import json
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from config import user, password
import psycopg2
```

```python
sdt = dt.datetime(1995, 1, 1)
edt = dt.datetime(2019, 11, 1)
#cpi=consumer price index 
```

```python
#Consumer Price Index for All Urban Consumers: All Items in U.S. City Average (CPIAUNCS)
CPI = web.DataReader("CPIAUCNS", "fred", sdt, edt)
 
CPI.rename(columns={"CPIAUCNS":"CPI"}, inplace=True)
```

```python
#Moody's Seasoned Aaa Corporate Bond Yield
aaa = web.DataReader("AAA", "fred", sdt, edt)
 
aaa.head()
```

```python
cpi_aaa_df = pd.merge(CPI,aaa, on="DATE", how ="left")
cpi_aaa_df
```

```python
#Moody's Seasoned Baa Corporate Bond Yield
baa = web.DataReader("BAA", "fred", sdt, edt)
 
baa.head()
```

```python
aaa_baa_df = pd.merge(aaa,baa, on="DATE", how ="left")
aaa_baa_df
```

```python
#10-Year Treasury Constant Maturity Minus 2-Year Treasury Constant Maturity
T10Y2YM = web.DataReader("T10Y2YM", "fred", sdt, edt)
T10Y2YM.rename(columns={"T10Y2YM":"10Minus2"}, inplace=True) 
T10Y2YM.head()
```

```python
#3-Month Treasury Bill: Secondary Market Rate
DTB3 = web.DataReader("DTB3", "fred", sdt, edt)

DTB3.head()
DTB3_C = DTB3.dropna()
 
```

```python
DTB3_M = DTB3_C.resample('1M').mean()
DTB3_M.rename(columns={"DTB3":"3monthBill"}, inplace=True)
print(DTB3_M)
```

```python
#SPY: an exchange-traded fund
SPY = web.get_data_yahoo('^GSPC', start=dt.datetime(1994, 1, 1), end=dt.datetime(2019, 12, 1))


SPY = SPY.drop(["High", "Low", "Open","Volume", "Adj Close"], axis=1)
SPY = SPY.resample('1M').mean()

SPY_Pct_Chg = SPY.pct_change(12)
SPY_Pct_Chg.head(20)

SPY_Merged = pd.merge(SPY,SPY_Pct_Chg, on="Date", how ="left")
```

```python
SPY_Merged.head()
```

```python
SPY_Merged.rename(columns = {'Close_x':'SPYClose', 'Close_y':'SPY12mFR'}, inplace=True) 
```

```python
SPY_Merged.head()
```

```python
SPY_Merged_C = SPY_Merged.dropna()
SPY_Merged_C.index.names = ['DATE']
SPY_Merged_C.head()
```

```python
rds_connection_string = user + ":" + password + "@localhost:5432/ETLStock_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
aaa_baa_df.to_sql(name='Bonds', con=engine, if_exists='append', index=True)
```

```python
CPI.to_sql(name='CPI', con=engine, if_exists='append', index=True)
```

```python
DTB3_M.to_sql(name='3monthTBill', con=engine, if_exists='append', index=True)
```

```python
T10Y2YM.to_sql(name='10Minus2', con=engine, if_exists='append', index=True)
```

```python
SPY_Merged_C.to_sql(name='S&PData', con=engine, if_exists='append', index=True)
```
